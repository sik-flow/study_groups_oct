## Lambda Functions

Lambda functions are often a convenient way to write throw-away functions on the fly. If you need to write a more complicated function you may still need to use the more formal def method, but lambda functions provide a quick and concise way to write functions.

```python
df['Length_of_Ticket_Lambda'] = df['Ticket'].map(lambda x: len(x))
```

```python
df.head()
```

```python
df.describe()
```

```python
# Drop Length_of_Ticket
df.drop(['Length_of_Ticket', 'Length_of_Ticket_Lambda'])
```

```python
df.drop(['Length_of_Ticket', 'Length_of_Ticket_Lambda'], axis = 1)
```

```python
df.drop(['Length_of_Ticket', 'Length_of_Ticket_Lambda'], axis = 1, inplace = True)
```

### Groupby

Use groupby methods to aggregate different groups in a dataframe

```python
# average age by Pclass
df.groupby('Pclass')['Age'].mean()
```

```python
# average ticket price by pclass
df.groupby('Pclass')['Fare'].mean()
```

```python
# average age by survived
df.groupby('Survived')['Age'].mean()
```

```python
# min, max, sum
df.groupby('Survived')['Fare'].min()
```

```python
df.groupby('Survived')['Fare'].max()
```

```python
df.groupby('Survived')['Fare'].sum()
```
